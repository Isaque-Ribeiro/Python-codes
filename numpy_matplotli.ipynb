```python
# FUNÇÕES SENO E COSSENO PELA SÉRIE DE TAYLOR
import numpy as n
from scipy.special import factorial
from math import pi

def cos(x):
  positivo = n.arange(0,50,4)
  negativo = n.arange(2,50,4)
  fatPst = factorial(positivo)
  fatNgt = factorial(negativo)
  termosPst = (x**positivo)/fatPst
  termosNgt = (x**negativo)/fatNgt
  a = termosPst.sum()
  b = termosNgt.sum()
  c = a - b
  return c


def sen(x):
  a = (1 - (cos(x))**2)**(1/2)
  return a
z = pi/6
print(cos(z))
print(sen(z))
print((sen(z)**2)+(cos(z)**2))
```

```python
# COMO COLOCAR PONTOS NA FUNÇÃO DE UM CÍRCULO
import numpy as n
import matplotlib.pyplot as p
from math import pi

f = 2*pi
abscissa = n.linspace(0,f,200)
ordenada = n.linspace(0,f,200)
x = (n.cos(abscissa))*3 + 3
y = (n.sin(ordenada))*3 + 3

p.scatter(x,y)
```

```python
import numpy as n
import matplotlib.pyplot as p

entrada = int(input('Dê o número de porntos para o cálculo da aproximação de pi: '))
x = n.linspace(0,100,(entrada))*0.01
y = (1 - x**2)**(1/2)
p.plot(x,y, 'black')

x = n.random.rand(1, entrada)
z = n.random.rand(1,entrada)
delta = z - y
dentro = delta < 0
exemplos = delta[dentro]
qnt = exemplos.size
pi = 4*(qnt/entrada)
print(pi)

x = x[0]
z = z[0]
xdentro = []
zdentro = []
xfora = []
zfora = []
for i in range (0, entrada):
  if ((x[i]**2 + z[i]**2) < 1):
    xdentro.append(x[i])
    zdentro.append(z[i])
  else:
    xfora.append(x[i])
    zfora.append(z[i])
xdentro = n.array(xdentro)
zdentro = n.array(zdentro)
xfora = n.array(xfora)
zfora = n.array(zfora)
p.plot(xdentro,zdentro, 'ro')
p.plot(xfora,zfora, 'bo')
p.show()
```

```python
import numpy as n


a = []
b = int(input('Qnt: '))
print('Dados:')
for i in range(0, b):
  c = int(input(''))
  a.append(c)

lin = int(input('qnt linhas: '))
col = int(input('qnt colunas: '))

mt = n.array(a)
print(mt)
mt = mt.reshape(lin,col)

media = mt.mean(1)
print(f'medias por linha: {media.reshape(col,1)}')
mediaT = mt.mean()
print(f'media total: {mediaT}')


SQTr = (media - mediaT)**2
SQTr = (SQTr.sum())*col
print(f'SQTr: {SQTr}')

SQT = (mt-mediaT)**2
print(f'SQT: {SQT.sum()}')

SQE = SQT.sum() - SQTr.sum()
print(f'SQE: {SQE}')

QMTr = SQTr/(col-1)
print(f'QMTr: {QMTr}')
QME = SQE/(lin*(col-1))
print(f'QME: {QME}')

F = QMTr/QME
print(f'F de Snedecor: {F}')
```
